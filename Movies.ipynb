```python
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
movies = pd.read_csv('movies.csv', encoding='latin1')

#we are only interested in US movies
movies = movies.drop(movies.loc[movies.country != 'USA'].index)

#find and drop rows that dont have both year and month info
movies = movies.drop(movies.loc[movies.released.str.len() < 7].index)

movies['genre'].value_counts()

movies.sort_values('runtime', ascending = False)
```

```python
movies['released_month'] = movies['released'].apply(lambda x: pd.to_datetime(x).month)
all_genre = movies.groupby(['genre', 'released_month'])['gross'].mean()
all_genre.plot('bar', figsize = (100, 15), fontsize = 20)
```

```python
by_year = movies.groupby(['genre', 'year', 'released_month'])['gross'].mean()
by_year.plot('bar', figsize = (500, 15), fontsize = 20)
```

```python
crime = movies.loc[movies['genre'] == 'Crime']
crime_by_month = crime.groupby('released_month')['gross'].mean()
crime_by_month.plot('bar',title='Crime')
crime_by_month.plot()
```
