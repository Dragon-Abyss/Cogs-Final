# Data Cleaning

Describe your data cleaning steps here.

```python
movies = pd.read_csv('movies.csv', encoding = "ISO-8859-1")

#find and drop rows that dont have both year and month info
movies = movies.drop(movies.loc[movies.released.str.len() < 7].index)

#we are only interested in US movies
movies = movies.drop(movies.loc[movies.country != 'USA'].index)

#make a YYYY-MM string to help group by year and month
def getYearMonth(s):
  return s.split("-")[0]+"-"+s.split("-")[1]

yearmonth = []

for index, row in movies.iterrows():
    yearmonth.append(getYearMonth(row['released']))

movies['year-month released'] = yearmonth

#examining the genres available in this dataframe
genre_counts = movies['genre'].value_counts()
print(genre_counts)

#comedy genre
comedy = movies.loc[(movies['genre'] == 'Comedy')]
comedy = comedy[['gross', 'year-month released']]
comedy_gross = comedy.groupby(['year-month released'])['gross'].mean()
comedy_gross.plot('bar', figsize=(100,15), fontsize=20)

len(movies)
```
